## Rule-based Chatbot using regex

Referensi tambahan:
1. https://www.w3schools.com/python/python_regex.asp
2. https://docs.python.org/3/library/re.html

```python
import re

JAWABAN_DEFINISI = """\
Chatbot adalah sistem yang berinteraksi lewat bahasa alami (teks/suara) untuk menjawab pertanyaan atau menjalankan tugas.
Komponen umum:
1) NLU (intent & entity),
2) Dialog/Policy (alur percakapan),
3) NLG (membentuk respons),
4) Integrasi kanal & backend (API/DB)."""

JAWABAN_KURIKULUM = """\
Kurikulum belajar chatbot:
1) Dasar NLP: tokenisasi, intent/entity.
2) Dialog management.
3) Retrieval (RAG bila perlu).
4) Prompting & evaluasi.
5) Keamanan & privasi.
6) Deployment & observability."""

JAWABAN_IMPLEMENTASI = """\
Langkah implementasi:
1) Kumpulkan knowledge base.
2) Pilih pendekatan (regex/rule, intent ML, atau LLM+RAG).
3) Buat pipeline: parse → tentukan intent → susun jawaban.
4) Tambah guardrails & fallback.
5) Deploy via FastAPI/Flask + integrasi UI chat/WA/Slack."""

JAWABAN_LIBRARY = """\
Library populer:
- Regex/rule: modul `re` Python.
- NLP klasik: scikit-learn, spaCy.
- LLM & RAG: OpenAI API, HuggingFace Transformers, FAISS/Chroma.
- Framework dialog: Rasa, LangChain/LlamaIndex.
- Serving: FastAPI/Flask."""

JAWABAN_RAG_VS_REGEX = """\
RAG vs Regex:
- Regex cocok untuk pola tetap & routing intent.
- RAG cocok saat jawaban perlu dicari dari dokumen dinamis.
- Keduanya bisa dipakai bareng: regex untuk ekstraksi, RAG/LLM untuk sintesis jawaban natural."""


rules = [
    (re.compile(r".*\b(apa itu|definisi|pengertian)\b.*\b(chat ?bot)\b.*", re.I), JAWABAN_DEFINISI),
    (re.compile(r".*\b(apa|hal|materi|yang)\b.*\b(dipelajari|belajar|kurikulum|roadmap)\b.*\b(chat ?bot)\b.*", re.I), JAWABAN_KURIKULUM),
    (re.compile(r".*\b(bagaimana|gimana|cara)\b.*\b(membuat|implementasi|membangun)\b.*\b(chat ?bot)\b.*", re.I), JAWABAN_IMPLEMENTASI),
    (re.compile(r".*\b(library|framework|tools?|paket)\b.*\b(chat ?bot|rag|nlp)\b.*", re.I), JAWABAN_LIBRARY),
    (re.compile(r".*\b(rag)\b.*\b(vs|atau|dibanding|bandingkan)\b.*\b(regex)\b.*", re.I), JAWABAN_RAG_VS_REGEX),
]

fallback = "Aku kurang paham pertanyaanmu. Coba tanya: 'apa itu chatbot', 'apa yang dipelajari untuk chatbot', 'bagaimana implementasi chatbot', atau 'library chatbot'."

def find_answer(text: str) -> str:
    for pat, ans in rules:
        if pat.match(text):
            return ans
    return fallback


if __name__ == "__main__":
    print(" Chatbot Belajar Chatbot. Ketik 'exit' untuk keluar.")
    while True:
        user = input("\nYou: ").strip()
        if not user or user.lower() in ("exit", "quit", "keluar"):
            print("Bot: Sampai jumpa!")
            break
        print("Bot:", find_answer(user))
```

## Rule-based Chatbot using NLTK
Sumber: https://www.analyticsvidhya.com/blog/2021/07/build-a-simple-chatbot-using-python-and-nltk/

```python
import nltk
from nltk.chat.util import Chat, reflections

# Define some patterns and responses
patterns = [
    [
        r"my name is (.*), i want eat (.*)",
        ["Hello %1, How are you today ?",]
    ],
    [
        r"hi|hey|hello",
        ["Hello", "Hey there",]
    ],
    [
        r"what is your name ?",
        ["I am a bot created for this course, you can call me anything!",]
    ],
    [
        r"how are you ?",
        ["I'm doing good, How about You ?",]
    ],
    [
        r"sorry (.*)",
        ["Its alright","Its OK, never mind",]
    ],
    [
        r"I am fine",
        ["Great to hear that, How can I help you?",]
    ],
    [
        r"i'm (.*) doing good",
        ["Nice to hear that","How can I help you?:)",]
    ],
    [
        r"(.*) age?",
        ["I'm a computer program duden. Seriously you are asking me this?",]
    ],
    [
        r"what (.*) want ?",
        ["Make me an offer I can't refuse",]
    ],
    [
        r"how (.*) health(.*)",
        ["I'm a computer program, so I'm always healthy ",]
    ],
    [
        r"(.*) (sports|game) ?",
        ["I'm a very big fan of Football",]
    ],
    [
        r".* i'm (depressed|sad) .*",
        ["I'm sorry to hear that",]
    ],
    [
        r".* all .*",
        ["In what way",]
    ],
    [
        r".* always .*",
        ["Can you think of a specific example",]
    ],
    [
        r"quit",
        ["BBye take care. See you soon :) ","It was nice talking to you. See you soon :)"]
    ],
]

# Create a chatbot
chatbot = Chat(patterns, reflections)

def main():
    print("Hello! I'm a simple chatbot. You can start chatting. Type 'quit' to exit.")
    while True:
        user_input = input("You: ")
        response = chatbot.respond(user_input)
        print("Bot:", response)
        if user_input.lower() == 'quit':
            break

if __name__ == "__main__":
    main()
```

## Chatbot WITH LLM

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

MODEL_ID = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

# Load
tok = AutoTokenizer.from_pretrained(MODEL_ID)
pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(
        MODEL_ID,
        torch_dtype="auto",
        device_map="auto"
    ),
    tokenizer=tok
)

def chat_once(user_msg: str) -> str:
    messages = [
        {"role": "system", "content": "You are an assistant who helps for medical"},
        {"role": "user", "content": user_msg},
    ]
    prompt = tok.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    out = pipe(
        prompt,
        max_new_tokens=256,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        eos_token_id=tok.eos_token_id,
    )[0]["generated_text"]

    reply = out[len(prompt):].strip()
    return reply

if __name__ == "__main__":
    print("HF Chatbot (TinyLlama). ketik 'exit' untuk keluar.")
    while True:
        u = input("You: ")
        if u.strip().lower() == "exit":
            break
        print("Bot:", chat_once(u))
```

# Chatbot RAG -LLM

Chatbot ini menggunakan dokumen PDF cerita fiksi “Sepuluh Sahabat ke Kota Senja”.
Secara garis besar, alurnya sebagai berikut:

1. Parsing PDF  → Ekstraksi teks halaman memakai pypdf.PdfReader. Teks dibersihkan (ligature, soft-hyphen, pemotongan “-\n”, spasi berlebih). Dokumen kemudian dipecah berdasarkan heading bernomor (jika ada); bila tidak terdeteksi, dipotong per ukuran tetap (default chunk_size=1200, overlap=200).


2. Struktur & Booster Domain Cerita → Konten diperkaya dengan booster pencarian untuk kode & istilah cerita:
C01–C10 (karakter), L01–L10 (lokasi), S1–S8 (adegan), K1–K4 (konflik), T01/T02 (tiket), R01/R02 (nota), P01 (peta), nama tokoh/lokasi (Aksa, Bela, Museum Riwayat Batu, Alun-Alun, Gedung Pusaka Angin, Danau Cermin, Pasar Malam Senja, Rumah Makan Panyawara, Terminal Caraka), serta jam penting 07:30, 09:00, 13:00. Booster ini meningkatkan skor kandidat yang relevan saat retrieve.

3. Embedding & Indexing (RAG) → Potongan teks diubah menjadi vektor menggunakan SentenceTransformer (BAAI/bge-m3) dengan normalisasi, lalu diindeks ke FAISS (IndexFlatIP) untuk semantic search. Parameter utama: k=12 untuk pencarian awal (top-k).

4. Re-ranking (opsional) → Hasil retrieve awal dapat diurut ulang oleh CrossEncoder (BAAI/bge-reranker-v2-m3) agar konteks paling relevan naik ke atas (default topk_final=6). Fitur ini bisa dimatikan lewat konfigurasi jika RAM terbatas.

5. LLM Generation → Jawaban disintesis oleh Qwen/Qwen2.5-1.5B-Instruct. Prompt dibentuk di fungsi make_messages, menegaskan aturan: jawab singkat, jelas, dan HANYA dari konteks. Jika tidak ada informasi relevan, bot menjawab: “Tidak ditemukan dalam naskah.” (tanpa menyertakan rujukan).

Fungsi kunci:

build_docs() → baca PDF, bersihkan teks, pecah jadi potongan.
retrieve() → FAISS search + booster kode/istilah cerita (+ opsional re-ranking).
make_messages() → susun prompt berisi konteks terpilih.
generate_answer() → panggil LLM untuk menghasilkan jawaban.
chat(query) → pipeline ujung-ke-ujung (retrieve → generate).

```python
# Cell 1 - Import libraries
import os, re, textwrap
from dataclasses import dataclass
from typing import List, Dict, Tuple
import numpy as np

from pypdf import PdfReader  # ekstraksi teks PDF berbasis teks

import faiss
from sentence_transformers import SentenceTransformer, CrossEncoder  # biarkan jika pakai re-ranking

import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

print("Libraries diimport")
```

```python
# Cell 2 - Setup PDF reader (Colab) — PDF-only tanpa OCR
import os, re
from pypdf import PdfReader

# Ganti ke path PDF kamu di Colab
PDF_PATH = "/content/Sepuluh Sahabat ke Kota Senja.pdf"

def verify_pdf(path: str) -> bool:
    if not os.path.isfile(path):
        raise FileNotFoundError(f"Tidak menemukan PDF di: {path}")
    try:
        reader = PdfReader(path)
        n_pages = len(reader.pages)
        if n_pages == 0:
            raise ValueError("PDF tidak memiliki halaman.")
        print(f"PDF OK — jumlah halaman: {n_pages}")
        return True
    except Exception as e:
        raise RuntimeError(f"Gagal membaca PDF: {e}")

def extract_pages(path: str):
    """Ekstrak teks per halaman (tanpa OCR). Halaman kosong di-skip."""
    reader = PdfReader(path)
    pages = []
    for i, page in enumerate(reader.pages, start=1):
        txt = page.extract_text() or ""
        # Normalisasi ringan
        txt = txt.replace("\u00ad", "")     # soft hyphen
        txt = txt.replace("-\n", "")        # sambung kata yang terpotong
        txt = re.sub(r"\s+", " ", txt).strip()
        if txt:
            pages.append({"page": i, "text": txt})
    print(f"Halaman dengan teks: {len(pages)}")
    return pages

# Verifikasi & contoh ekstraksi
verify_pdf(PDF_PATH)
pages_text = extract_pages(PDF_PATH)
if pages_text:
    print("Cuplikan halaman 1 (200 karakter):")
    print(pages_text[0]["text"][:200])
```

```python
# Cell 4 - PDF processing (tanpa OCR)

def clean_text(s: str) -> str:
    """Bersihkan teks dari karakter aneh & normalisasi spasi/baris."""
    s = (s or "")
    s = s.replace("\x00", " ").replace("\ufb01", "fi").replace("\ufb02", "fl")  # ligatures umum
    s = s.replace("\u00ad", "")        # soft hyphen
    s = s.replace("-\n", "")           # sambung kata terpotong di line break
    s = re.sub(r"\n{3,}", "\n\n", s)   # kompres newline berlebih
    s = re.sub(r"[ \t]+", " ", s)      # kompres spasi/tab
    return s.strip()

def read_pdf_text(path: str, min_chars: int = 50) -> Tuple[str, List[Tuple[int, str]]]:
    print(f"Membaca PDF: {os.path.basename(path)}")

    reader = PdfReader(path)
    pages: List[Tuple[int, str]] = []

    for i, p in enumerate(reader.pages, start=1):
        raw = p.extract_text() or ""
        t = clean_text(raw)
        pages.append((i, t))
        print(f"  Halaman {i}: {len(t)} karakter")

    # Peringatan untuk halaman yang tampak kosong/scan
    suspect = [i for (i, t) in pages if len(t) < min_chars]
    if suspect:
        print(f"Peringatan: {len(suspect)} halaman terdeteksi minim teks {suspect}. "
              f"Kemungkinan berbasis gambar. Mode ini PDF-only, OCR dinonaktifkan.")

    full = "\n\n".join([f"[page {i}]\n{t}" for i, t in pages])
    total_chars = sum(len(t) for _, t in pages)
    print(f"Total: {len(pages)} halaman, {total_chars} karakter (tanpa OCR)")

    return full, pages

# Contoh pakai (opsional):
full_text, pages_list = read_pdf_text(PDF_PATH)
print(full_text[:500])
```

```python
# Cell 5 - Path file PDF (PDF-only)
import os

# Ganti ke path PDF kamu (Colab / local)
PDF_PATH = "/content/Sepuluh Sahabat ke Kota Senja.pdf"  # ubah jika perlu

def set_pdf_list(path: str):
    if os.path.isfile(path):
        print(f"File PDF ditemukan: {os.path.basename(path)}")
        return [path]
    else:
        print(f"File tidak ditemukan: {path}")
        print("Pastikan path file sudah benar.")
        return []

PDFS = set_pdf_list(PDF_PATH)
```

```python
# Cell 6 - Load & proses dokumen PDF (Kota Senja, PDF-only)
import re, os

all_docs = []

# Pola regex sesuai struktur naskah "Kota Senja"
# 1) Heading bernomor: "1. Daftar Karakter (C01–C10)", "2. Daftar Lokasi (L01–L10)", dst.
RE_HEADING_NUM = re.compile(
    r'(?m)^\s*(\d{1,2}\.\s+[^\n]+)\s*\n(.*?)(?=^\s*\d{1,2}\.\s+[^\n]+|\Z)',
    flags=re.DOTALL

)

# 2) Adegan: "S1 — Titik Kumpul di L01", "S2 — Perjalanan ke L02", dst.
#    Mendukung variasi dash: -, –, —
RE_SCENE = re.compile(
    r'(?m)^\s*(S\d{1,2}\s*[—\-–]\s*[^\n]+)\s*\n(.*?)(?=^\s*S\d{1,2}\s*[—\-–]\s*[^\n]+|\Z)',
    flags=re.DOTALL
)

# 3) Q–A: baris yang diawali "Q:" lalu diikuti "A:" (bisa berulang)
RE_QA_PAIR = re.compile(
    r'(?m)^\s*Q\s*:\s*(.+?)\s*\n\s*A\s*:\s*(.+?)(?=\n\s*Q\s*:|\Z)',
    flags=re.DOTALL
)

for path in PDFS:
    print(f"\n=== Proses: {os.path.basename(path)} ===")
    full_text, page_texts = read_pdf_text(path)  # fungsi dari Cell 4
    source_name = os.path.basename(path)
    before_count = len(all_docs)

    # 1) Ekstrak per heading bernomor
    for m in RE_HEADING_NUM.finditer(full_text):
        title = m.group(1).strip()
        body  = (m.group(2) or "").strip()

        # hint halaman: cari kemunculan judul
        page_hint = None
        pat = re.escape(title)
        for (pg, txt) in page_texts:
            if re.search(pat, txt, flags=re.IGNORECASE):
                page_hint = pg
                break

        all_docs.append({
            "text": f"{title}\n{body}",
            "section": title,
            "source": source_name,
            "page_hint": page_hint
        })

    # 2) Ekstrak Adegan Sx — ... (potongan lebih granular)
    for m in RE_SCENE.finditer(full_text):
        title = m.group(1).strip()      # contoh: "S3 — Museum L02 (09:00–10:00)"
        body  = (m.group(2) or "").strip()

        # hint halaman: cukup cari "S\d+"-nya
        page_hint = None
        s_code = re.match(r'^\s*(S\d{1,2})', title)
        if s_code:
            s_pat = re.escape(s_code.group(1))
            for (pg, txt) in page_texts:
                if re.search(rf'\b{s_pat}\b', txt):
                    page_hint = pg
                    break

        all_docs.append({
            "text": f"{title}\n{body}",
            "section": title,            # simpan judul adegan sebagai section
            "source": source_name,
            "page_hint": page_hint
        })

    # 3) Ekstrak Q–A (berguna untuk retrieval pertanyaan langsung)
    for m in RE_QA_PAIR.finditer(full_text):
        q = m.group(1).strip()
        a = m.group(2).strip()
        qa_text = f"Q: {q}\nA: {a}"
        all_docs.append({
            "text": qa_text,
            "section": "Q-A",
            "source": source_name,
            "page_hint": None
        })

    # 4) Fallback: jika tidak ada potongan yang berhasil ditambahkan dari file ini,
    #    gunakan chunking ukuran tetap
    after_count = len(all_docs)
    if after_count == before_count:
        print("Pola heading/adegan/Q-A tidak ditemukan. Menggunakan fallback chunking...")
        step = 1200
        for i in range(0, len(full_text), step):
            chunk = full_text[i:i+step].strip()
            if not chunk:
                continue
            all_docs.append({
                "text": chunk,
                "section": "CHUNK",
                "source": source_name,
                "page_hint": None
            })

print("chunks:", len(all_docs))
```

```python
# Cell 7 - Setup embedding & FAISS index (PDF-only)
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

if not all_docs:
    raise ValueError("Tidak ada dokumen untuk di-embed. Pastikan Cell 6 menghasilkan 'all_docs' > 0.")

EMB_MODEL = "BAAI/bge-m3"
print(f"Memuat embedding model: {EMB_MODEL}")
embedder = SentenceTransformer(EMB_MODEL)

corpus = [d["text"] for d in all_docs]
print(f"Jumlah chunk untuk di-embed: {len(corpus)}")

# Encode ke vektor (cosine via normalize_embeddings=True → pakai IndexFlatIP)
X = embedder.encode(
    corpus,
    batch_size=64,
    normalize_embeddings=True,
    show_progress_bar=True
)
X = np.array(X, dtype="float32")

print("Membangun FAISS index (Inner Product / Cosine-normalized)...")
index = faiss.IndexFlatIP(X.shape[1])
index.add(X)

print(f"Index siap. Dimensi: {X.shape[1]}, Jumlah vektor: {index.ntotal}")
```

```python
# Cell 8 - Setup reranker (opsional tapi disarankan)
from sentence_transformers import CrossEncoder
import numpy as np

RERANKER = "BAAI/bge-reranker-v2-m3"
print(f"Memuat reranker: {RERANKER}")
reranker = CrossEncoder(RERANKER)  # otomatis pakai GPU jika tersedia (via torch)

def rerank_pairs(query: str, docs: List[Dict], topk: int = 5) -> List[Dict]:
    """
    Re-ranking doc hasil retrieval awal (mis. dari FAISS) dengan CrossEncoder.
    Mengembalikan hingga topk dokumen terbaik, menambahkan 'rerank_score'.
    """
    if not docs:
        return []
    k = min(topk, len(docs))
    pairs = [[query, d["text"]] for d in docs]
    scores = np.asarray(reranker.predict(pairs), dtype="float32")
    order = np.argsort(-scores)[:k]
    return [docs[i] | {"rerank_score": float(scores[i])} for i in order]
```

```python
# Cell 9 - Setup LLM
LLM = "Qwen/Qwen2.5-1.5B-Instruct"

device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained(LLM, use_fast=True)
model = AutoModelForCausalLM.from_pretrained(
    LLM,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    low_cpu_mem_usage=True,
)
model.to(device)
model.eval()

def make_messages(query: str, ctxs: List[Dict]) -> List[Dict]:
    bullets = []
    for c in ctxs:
        snippet = re.sub(r"\s{2,}", " ", c["text"]).strip()[:900]
        bullets.append(f"- {snippet}")
    context_block = "\n".join(bullets) if bullets else "(konteks kosong)"
    return [
        {
            "role": "system",
            "content": (
                "Anda asisten akademik yang menjawab berdasarkan konteks yang diberikan. "
                "Jawab SINGKAT, JELAS, dan HANYA dari konteks. "
                "Jika tidak ada informasi relevan, balas: 'Tidak ditemukan dalam dokumen.' "
                "Jangan sertakan rujukan/sumber dalam jawaban."
            ),
        },
        {
            "role": "user",
            "content": (
                f"KONTEKS:\n{context_block}\n\n"
                f"PERTANYAAN:\n{query}\n\n"
                f"Jawab singkat dan jelas tanpa menyertakan rujukan."
            ),
        },
    ]

@torch.inference_mode()
def generate_answer(messages, max_new_tokens=256):
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    input_len = inputs["input_ids"].shape[1]
    out = model.generate(
        **inputs,
        do_sample=False,
        max_new_tokens=max_new_tokens,
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id,
    )[0]
    gen = out[input_len:]
    txt = tokenizer.decode(gen, skip_special_tokens=True).strip()
    return re.sub(r"^assistant\s*:\s*", "", txt, flags=re.I) or "Tidak ditemukan dalam dokumen."
```

```python
# Cell 10 - Retrieval & Chat khusus cerita "Kota Senja"

from dataclasses import dataclass
from typing import List, Dict, Optional
import numpy as np
import re, os

@dataclass
class Retrieved:
    text: str
    score: float
    section: str
    source: str
    page_hint: Optional[int]

# --- Booster domain "Kota Senja" ---
# Pola kode: C01–C10, L01–L10, S1–S8, K1–K4, T01/T02, R01/R02, P01
PAT_CODE = re.compile(r"\b((?:C|L|S|K|T|R|P)\d{1,2})\b", re.IGNORECASE)

# Nama karakter & lokasi utama + istilah perjalanan dari naskah
STORY_TERMS = (
    r"(Aksa|Bela|Ciko|Dira|Eka|Fajar|Gita|Hani|Iwan|Juno|"
    r"Stasiun Mentari|Museum Riwayat Batu|Alun-?Alun( Kota Senja)?|Gedung Pusaka Angin|"
    r"Danau Cermin|Pasar Malam Senja|Rumah Makan Panyawara|Terminal Caraka|Kota Senja|"
    r"timekeeper|P3K|artefak|tiket|nota|peta|buffer waktu|itinerary|timeline|sesi foto|"
    r"07:30|09:00|13:00)"
)

IMPORTANT_TERMS = rf"({STORY_TERMS}|{PAT_CODE.pattern})"

def _boost_ids_from_query(query: str) -> List[int]:
    """Ambil doc yang mengandung kode (Cxx/Lxx/Sx/Kx/Txx/Rxx/Pxx) atau nama penting dari query."""
    boost_ids: List[int] = []
    if not query:
        return boost_ids

    # Kode eksplisit dalam query (mis. 'L02', 'C04', 'R02')
    codes = set(m.group(1).upper() for m in PAT_CODE.finditer(query))

    # Nama/keyword penting (huruf besar untuk pencocokan case-insensitive)
    names = [
        "AKSA","BELA","CIKO","DIRA","EKA","FAJAR","GITA","HANI","IWAN","JUNO",
        "STASIUN MENTARI","MUSEUM RIWAYAT BATU","ALUN","GEDUNG PUSAKA ANGIN",
        "DANAU CERMIN","PASAR MALAM SENJA","RUMAH MAKAN PANYAWARA","TERMINAL CARAKA","KOTA SENJA",
        "07:30","09:00","13:00"
    ]

    for i, d in enumerate(all_docs):
        sec = (d.get("section") or "")
        txt = (d.get("text") or "")
        hit = False
        for c in codes:
            if re.search(rf"\b{re.escape(c)}\b", sec, flags=re.I) or re.search(rf"\b{re.escape(c)}\b", txt, flags=re.I):
                hit = True; break
        if not hit:
            for nm in names:
                if re.search(nm, sec, flags=re.I) or re.search(nm, txt, flags=re.I):
                    hit = True; break
        if hit:
            boost_ids.append(i)
    return boost_ids

def make_messages_story(query: str, ctxs: List[Dict]) -> List[Dict]:
    """Template pesan untuk LLM yang menjawab soal cerita 'Kota Senja'."""
    bullets = []
    for c in ctxs:
        snippet = re.sub(r"\s{2,}", " ", c["text"]).strip()[:900]
        bullets.append(f"- {snippet}")
    context_block = "\n".join(bullets) if bullets else "(konteks kosong)"
    return [
        {
            "role": "system",
            "content": (
                "Anda adalah asisten untuk cerita fiksi 'Sepuluh Sahabat ke Kota Senja'. "
                "Jawab SINGKAT, JELAS, dan HANYA berdasarkan konteks. "
                "Jika tidak ada informasi relevan di konteks, balas: 'Tidak ditemukan dalam naskah.' "
                "Jangan menambahkan rujukan/sumber dalam jawaban."
            ),
        },
        {
            "role": "user",
            "content": (
                f"KONTEKS:\n{context_block}\n\n"
                f"PERTANYAAN:\n{query}\n\n"
                f"Jawab singkat dan jelas tanpa menyertakan rujukan."
            ),
        },
    ]

def retrieve(query: str, k: int = 12, topk_final: int = 6) -> List[Retrieved]:
    # 1) Booster dari kode/nama pada query
    bonus_ids = _boost_ids_from_query(query)

    # 2) Semantic top-k via FAISS
    qv = np.array(embedder.encode([query], normalize_embeddings=True), dtype="float32")
    D, I = index.search(qv, min(k, index.ntotal))
    I = I[0].tolist(); D = D[0].tolist()

    # 3) Gabungkan kandidat + skor awal
    cand, seen = [], set()
    for bid in bonus_ids + I:
        if bid in seen:
            continue
        seen.add(bid)
        d = all_docs[bid]
        base = D[I.index(bid)] if bid in I else 0.0
        sc = base
        if bid in bonus_ids:
            sc += 0.12  # bonus untuk kecocokan kode/nama dari query
        if re.search(IMPORTANT_TERMS, d["text"], flags=re.I):
            sc += 0.05  # bonus ringan untuk istilah penting naskah
        cand.append(d | {"score": float(sc)})

    if not cand:
        return []

    # 4) Re-rank (CrossEncoder) → ambil topk_final
    reranked = rerank_pairs(query, cand, topk=topk_final)
    out: List[Retrieved] = []
    for c in reranked:
        out.append(Retrieved(
            text=c["text"],
            score=float(c.get("rerank_score", c.get("score", 0.0))),
            section=c["section"],
            source=c["source"],
            page_hint=c["page_hint"]
        ))
    return out

def chat(query: str) -> Dict:
    ctxs = retrieve(query)
    msgs = make_messages_story(query, [c.__dict__ for c in ctxs])
    ans = generate_answer(msgs)
    cites = [{"section": c.section, "page_hint": c.page_hint, "source": c.source} for c in ctxs]
    if ans.strip().lower() == (query or "").strip().lower():
        ans = "Tidak ditemukan dalam naskah."
    return {"answer": ans, "contexts": cites}

print("="*60)
print("RAG 'Sepuluh Sahabat ke Kota Senja' — siap dipakai!")
print("="*60)
```

```python
# Cell 11 - Chat interaktif

def start_chat(show_sources: bool = False):
    """Mulai chat interaktif untuk naskah 'Sepuluh Sahabat ke Kota Senja'."""
    print("CHATBOT 'Kota Senja' AKTIF!")
    print("Tanya apa saja tentang naskah 'Sepuluh Sahabat ke Kota Senja'.")
    print("Ketik 'exit' untuk keluar.")
    print("="*50)

    while True:
        try:
            q = input("\nAnda: ").strip()
        except EOFError:
            print("\nSesi dihentikan.")
            break

        if not q or q.lower() in {"exit", "quit", "q"}:
            print("Terima kasih!")
            break

        out = chat(q)
        print(f"\nBot: {out['answer']}")

        if show_sources and out.get("contexts"):
            print("\n[Sumber konteks]")
            for i, c in enumerate(out["contexts"], start=1):
                sec = c.get("section") or "—"
                pg  = c.get("page_hint")
                src = c.get("source") or ""
                pg_str = f"hal {pg}" if pg else "hal ?"
                print(f"{i}. {sec} ({pg_str}, {src})")

        print("-" * 50)

if __name__ == "__main__":
    # Ubah ke True jika ingin melihat sumber konteks yang dipakai
    start_chat(show_sources=False)
```
