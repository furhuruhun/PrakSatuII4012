## Rule-based Chatbot using regex

Referensi tambahan:

```python
import re

ANS_BEASISWA = """\
Beasiswa bertujuan menumbuhkan potensi dan daya juang (Pasal 3).
Jenisnya terdiri dari Beasiswa Prestasi dan Beasiswa Ekonomi (Pasal 4).
Mahasiswa penerima beasiswa sarjana berkontribusi melalui kerja paruh waktu maksimal 2 jam/hari atau 10 jam/pekan (Pasal 5)."""

ANS_ASRAMA = """\
Asrama Mahasiswa ITB dapat dihuni selama-lamanya 1 tahun (Pasal 6).
Layanan asrama diprioritaskan untuk mahasiswa Tahap Persiapan Bersama (TPB).
Penghuni wajib mengikuti aturan tata tertib dan kegiatan pengembangan karakter (Pasal 6)."""

ANS_KESEHATAN = """\
Layanan kesehatan membantu mahasiswa menjaga kesehatan psikis dan mental selama masa studi (Pasal 7).
ITB memberikan subsidi biaya rawat jalan yang nilainya ditentukan oleh Unit Kerja layanan kesehatan (Pasal 7)."""

ANS_KONSELING = """\
Layanan bimbingan dan konseling mencakup bimbingan karir, proses belajar, hingga konseling masalah pribadi atau sosial (Pasal 8) .
Layanan ini didasarkan pada asas kerahasiaan dan kesukarelaan (Pasal 8)."""

ANS_DISABILITAS = """\
Layanan disabilitas bertujuan agar mahasiswa memaksimalkan potensi dan lulus dengan baik (Pasal 9).
ITB menyediakan serta meminjamkan sarana prasarana khusus bagi mahasiswa penyandang disabilitas (Pasal 9)"""

ANS_KARAKTER = """\
Pengembangan karakter bertujuan menanamkan jujur, kerja sama tim, dan kepekaan sosial (Pasal 10).
Kegiatannya meliputi Studium Generale, latihan kepemimpinan, hingga pengabdian masyarakat (Pasal 10)."""

ANS_KEWIRAUSAHAAN = """\
Pengembangan kewirausahaan mencakup pelatihan, perekrutan, dan kesempatan magang (Pasal 11).
Program ini dijalankan melalui jejaring kerja sama dengan industri dan pemerintah (Pasal 11)."""

ANS_IZIN_KEGIATAN = """\
Izin kegiatan tingkat ITB diberikan oleh Unit Kerja Kemahasiswaan, sedangkan tingkat Prodi/Fakultas diberikan oleh Dekan (Pasal 14).
Kegiatan harus sesuai harkat pendidikan dan mendapat persetujuan dosen pembimbing (Pasal 15)."""

ANS_ORGANISASI = """\
Organisasi kemahasiswaan tidak boleh berafiliasi dengan partai politik atau organisasi luar ITB (Pasal 17).
Jenisnya meliputi Kabinet KM ITB (terpusat), UKM (institut), serta Himpunan Mahasiswa (Fakultas/Prodi) (Pasal 18) ."""

ANS_MAPRES = """\
Syarat Mahasiswa Berprestasi (Mapres) tingkat Sarjana: Telah menempuh 6 semester, usia maksimal 22 tahun, memiliki prestasi internasional/nasional, serta IP tinggi (Pasal 27)."""

ANS_GANESHA = """\
Penghargaan Ganesha terdiri atas 4 kategori (Pasal 28) :
1. Ganesha Perkasa (Olahraga)
2. Ganesha Rasa (Seni/Budaya)
3. Ganesha Karsa (Ilmiah/Sosial)
4. Ganesha Karya (Inovatif)"""

rules = [
    (re.compile(r"(?=.*(beasiswa|bantuan biaya|dana))(?=.*(apa|gimana|cara|syarat|tujuan|tugas))", re.I), ANS_BEASISWA),
    (re.compile(r"(?=.*(asrama|tinggal|huni))(?=.*(berapa lama|durasi|siapa|tpb|syarat))", re.I), ANS_ASRAMA),
    (re.compile(r"(?=.*(sehat|sakit|dokter|rawat|medis))(?=.*(apa|biaya|subsidi|gimana))", re.I), ANS_KESEHATAN),
    (re.compile(r"(?=.*(konseling|curhat|psikolog|mental))(?=.*(apa|maksud|definisi|tujuan|gimana))", re.I), ANS_KONSELING),
    (re.compile(r"(?=.*(disabilitas|difabel|berkebutuhan khusus))(?=.*(apa|fasilitas|layanan|bantuan))", re.I), ANS_DISABILITAS),
    (re.compile(r"(?=.*(karakter|studium generale|sg|latihan pimpinan))(?=.*(apa|tujuan|kegiatan))", re.I), ANS_KARAKTER),
    (re.compile(r"(?=.*(wirausaha|bisnis|startup|magang))(?=.*(apa|gimana|program|kerjasama))", re.I), ANS_KEWIRAUSAHAAN),
    (re.compile(r"(?=.*(izin|boleh|prosedur))(?=.*(kegiatan|acara|event|prodi|fakultas))", re.I), ANS_IZIN_KEGIATAN),
    (re.compile(r"(?=.*(organisasi|himpunan|ukm|km|kabinet))(?=.*(jenis|macam|apa|larangan|politik))", re.I), ANS_ORGANISASI),
    (re.compile(r"(?=.*(mapres|mahasiswa berprestasi))(?=.*(apa|syarat|kriteria|daftar|siapa))", re.I), ANS_MAPRES),
    (re.compile(r"(?=.*(ganesha|penghargaan))(?=.*(apa|jenis|kategori|perkasa|rasa|karsa))", re.I), ANS_GANESHA),
]

fallback = "Maaf, Coba tanya hal lain kak, entah beasiswa, asrama, atau organisasi gitu."

def clean_text(text: str) -> str:
    """Bersihin input user: lowercase, hapus tanda baca, dan buang filler."""
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    fillers = r'\b(sih|itu|dong|lah|kah|ya|ini|itu|banget|deh|an)\b'
    text = re.sub(fillers, '', text)
    text = " ".join(text.split())
    return text

def find_answer(text: str) -> str:
    cleaned = clean_text(text)
    for pat, ans in rules:
        if pat.match(cleaned):
            return ans
    return fallback


if __name__ == "__main__":
    print(" AI Pengganti CS peraturan akademik. Ketik 'exit', 'quit', 'keluar' untuk keluar.")
    while True:
        user = input("\nYou: ").strip()
        if not user or user.lower() in ("exit", "quit", "keluar"):
            print("AI: Until next time!")
            break
        print("AI:", find_answer(user))
```

## Chatbot WITH LLM

```python
import os, warnings
from transformers.utils import logging
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

os.environ["HF_HUB_DISABLE_PROGRESS_BAR"] = "1"
warnings.filterwarnings("ignore")
logging.set_verbosity_error()

MODEL_ID = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

# Load
tok = AutoTokenizer.from_pretrained(MODEL_ID)
pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(
        MODEL_ID,
        torch_dtype="auto",
        device_map="auto"
    ),
    tokenizer=tok
)

def chat_once(user_msg: str) -> str:
    messages = [
        {
        "role": "system", "content":
        """
        Kamu adalah chatbot peraturan kemahasiswaan ITB.
        Gunakan hanya informasi berikut untuk menjawab.

        FAQ:
        1. Layanan kemahasiswaan ITB meliputi beasiswa, asrama mahasiswa, layanan kesehatan mahasiswa, bimbingan konseling, pengembangan karakter, pengembangan keprofesian dan kewirausahaan, kegiatan kemahasiswaan, dan layanan bagi penyandang disabilitas.
        2. ITB dapat memberikan beasiswa prestasi dan beasiswa ekonomi.
        3. Mahasiswa dapat tinggal di asrama maksimal 1 tahun.
        4. Penghargaan Ganesha terdiri dari Ganesha Perkasa, Ganesha Rasa, Ganesha Karsa, dan Ganesha Karya.
        5. Terdapat 4 jenis organisasi yang ada di ITB yaitu Keluarga Mahasiswa ITB, Unit Kegiatan Mahasiswa, Himpunan Mahasiswa Fakultas/Sekolah, dan Himpunan Mahasiswa Program Studi.
        6. Layanan bimbingan dan konseling diberikan atas permintaan siswa dan dapat dilakukan secara luring maupun daring (pada jam kerja).
        7. Mahasiswa Berprestasi dipilih berdasarkan IPK, prestasi akademik/non-akademik, tidak pernah melanggar aturan, usia maksimal 22 tahun, dan meksimal semester 6.
        8. UKM dapat didirikan dengan minimal 100 anggota yang bersal dari 3 program studi dari fakultas atau sekolah yang berbeda.
        9. Izin untuk kegiatan kemahasiswaan diberikan secara resmi oleh Dekan.
        10. Mahasiswa non-TPB ITB dapat tinggal di asrama selama memenuhi persyaratan administratif.

        Jika pertanyaan di luar informasi ini, jawab: "Maaf, informasi tersebut belum tersedia."
        """
        # "You are a helpful assistant that answers questions about Institut Teknologi Bandung student regulations and student services."
        #  """
        # You are an assistant for ITB student services.

        # Understand the question and reason in English internally,
        # but provide the final answer in Indonesian.

        # If you do not know the answer, say:
        # 'Maaf, saya tidak memiliki informasi tersebut.'
        # """
        },
        {"role": "user", "content": user_msg},
    ]
    prompt = tok.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    out = pipe(
        prompt,
        max_new_tokens=256,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        eos_token_id=tok.eos_token_id,
    )[0]["generated_text"]

    reply = out[len(prompt):].strip()
    return reply

if __name__ == "__main__":
    print("Halo! Aku chatbot Peraturan Kemahasiswaan ITB. Yuk tanya aku kalau kamu bingung soal aturan kampus.\nKetik 'quit' untuk keluar.\n")
    while True:
        # print("-----------------------------------------------------------------")
        u = input("You: ")
        if u.strip().lower() == "exit":
            break
        print("Bot:", chat_once(u))
        print("--------------------------------------------------------------------------------------------\n")
```

# Chatbot RAG -LLM

Chatbot ini menggunakan dokumen PDF cerita fiksi “Sepuluh Sahabat ke Kota Senja”.
Secara garis besar, alurnya sebagai berikut:

1. Parsing PDF  → Ekstraksi teks halaman memakai pypdf.PdfReader. Teks dibersihkan (ligature, soft-hyphen, pemotongan “-\n”, spasi berlebih). Dokumen kemudian dipecah berdasarkan heading bernomor (jika ada); bila tidak terdeteksi, dipotong per ukuran tetap (default chunk_size=1200, overlap=200).


2. Struktur & Booster Domain Cerita → Konten diperkaya dengan booster pencarian untuk kode & istilah cerita:
C01–C10 (karakter), L01–L10 (lokasi), S1–S8 (adegan), K1–K4 (konflik), T01/T02 (tiket), R01/R02 (nota), P01 (peta), nama tokoh/lokasi (Aksa, Bela, Museum Riwayat Batu, Alun-Alun, Gedung Pusaka Angin, Danau Cermin, Pasar Malam Senja, Rumah Makan Panyawara, Terminal Caraka), serta jam penting 07:30, 09:00, 13:00. Booster ini meningkatkan skor kandidat yang relevan saat retrieve.

3. Embedding & Indexing (RAG) → Potongan teks diubah menjadi vektor menggunakan SentenceTransformer (BAAI/bge-m3) dengan normalisasi, lalu diindeks ke FAISS (IndexFlatIP) untuk semantic search. Parameter utama: k=12 untuk pencarian awal (top-k).

4. Re-ranking (opsional) → Hasil retrieve awal dapat diurut ulang oleh CrossEncoder (BAAI/bge-reranker-v2-m3) agar konteks paling relevan naik ke atas (default topk_final=6). Fitur ini bisa dimatikan lewat konfigurasi jika RAM terbatas.

5. LLM Generation → Jawaban disintesis oleh Qwen/Qwen2.5-1.5B-Instruct. Prompt dibentuk di fungsi make_messages, menegaskan aturan: jawab singkat, jelas, dan HANYA dari konteks. Jika tidak ada informasi relevan, bot menjawab: “Tidak ditemukan dalam naskah.” (tanpa menyertakan rujukan).

Fungsi kunci:

build_docs() → baca PDF, bersihkan teks, pecah jadi potongan.
retrieve() → FAISS search + booster kode/istilah cerita (+ opsional re-ranking).
make_messages() → susun prompt berisi konteks terpilih.
generate_answer() → panggil LLM untuk menghasilkan jawaban.
chat(query) → pipeline ujung-ke-ujung (retrieve → generate).

```python
# Cell 2 - Setup PDF reader (Colab) — PDF-only tanpa OCR
import os, re
from pypdf import PdfReader

# Ganti ke path PDF kamu di Colab
PDF_PATH = "/content/Peraturan_Kemahasiswaan_2022.pdf"

def verify_pdf(path: str) -> bool:
    if not os.path.isfile(path):
        raise FileNotFoundError(f"Tidak menemukan PDF di: {path}")
    try:
        reader = PdfReader(path)
        n_pages = len(reader.pages)
        if n_pages == 0:
            raise ValueError("PDF tidak memiliki halaman.")
        print(f"PDF OK — jumlah halaman: {n_pages}")
        return True
    except Exception as e:
        raise RuntimeError(f"Gagal membaca PDF: {e}")

def extract_pages(path: str):
    """Ekstrak teks per halaman (tanpa OCR). Halaman kosong di-skip."""
    reader = PdfReader(path)
    pages = []
    for i, page in enumerate(reader.pages, start=1):
        txt = page.extract_text() or ""
        # Normalisasi ringan
        txt = txt.replace("\u00ad", "")     # soft hyphen
        txt = txt.replace("-\n", "")        # sambung kata yang terpotong
        txt = re.sub(r"\s+", " ", txt).strip()
        if txt:
            pages.append({"page": i, "text": txt})
    print(f"Halaman dengan teks: {len(pages)}")
    return pages

# Verifikasi & contoh ekstraksi
verify_pdf(PDF_PATH)
pages_text = extract_pages(PDF_PATH)
if pages_text:
    print("Cuplikan halaman 1 (200 karakter):")
    print(pages_text[0]["text"][:200])
```

```python
# Cell 4 - PDF processing (tanpa OCR)

def clean_text(s: str) -> str:
    """Bersihkan teks dari karakter aneh & normalisasi spasi/baris."""
    s = (s or "")
    s = s.replace("\x00", " ").replace("\ufb01", "fi").replace("\ufb02", "fl")  # ligatures umum
    s = s.replace("\u00ad", "")        # soft hyphen
    s = s.replace("-\n", "")           # sambung kata terpotong di line break
    s = re.sub(r"\n{3,}", "\n\n", s)   # kompres newline berlebih
    s = re.sub(r"[ \t]+", " ", s)      # kompres spasi/tab
    return s.strip()

def read_pdf_text(path: str, min_chars: int = 50) -> Tuple[str, List[Tuple[int, str]]]:
    print(f"Membaca PDF: {os.path.basename(path)}")

    reader = PdfReader(path)
    pages: List[Tuple[int, str]] = []

    for i, p in enumerate(reader.pages, start=1):
        raw = p.extract_text() or ""
        t = clean_text(raw)
        pages.append((i, t))
        print(f"  Halaman {i}: {len(t)} karakter")

    # Peringatan untuk halaman yang tampak kosong/scan
    suspect = [i for (i, t) in pages if len(t) < min_chars]
    if suspect:
        print(f"Peringatan: {len(suspect)} halaman terdeteksi minim teks {suspect}. "
              f"Kemungkinan berbasis gambar. Mode ini PDF-only, OCR dinonaktifkan.")

    full = "\n\n".join([f"[page {i}]\n{t}" for i, t in pages])
    total_chars = sum(len(t) for _, t in pages)
    print(f"Total: {len(pages)} halaman, {total_chars} karakter (tanpa OCR)")

    return full, pages

# Contoh pakai (opsional):
full_text, pages_list = read_pdf_text(PDF_PATH)
print(full_text[:500])
```

```python
# Cell 5 - Path file PDF (PDF-only)
import os

# Ganti ke path PDF kamu (Colab / local)
PDF_PATH = "/content/Peraturan_Kemahasiswaan_2022.pdf"  # ubah jika perlu

def set_pdf_list(path: str):
    if os.path.isfile(path):
        print(f"File PDF ditemukan: {os.path.basename(path)}")
        return [path]
    else:
        print(f"File tidak ditemukan: {path}")
        print("Pastikan path file sudah benar.")
        return []

PDFS = set_pdf_list(PDF_PATH)
```

```python
# Cell 6 - Load & proses dokumen PDF (Kota Senja, PDF-only)
import re, os

all_docs = []

# Pola regex sesuai struktur naskah "Kota Senja"
# 1) Heading bernomor: "1. Daftar Karakter (C01–C10)", "2. Daftar Lokasi (L01–L10)", dst.
RE_HEADING_NUM = re.compile(
    r'(?m)^\s*(\d{1,2}\.\s+[^\n]+)\s*\n(.*?)(?=^\s*\d{1,2}\.\s+[^\n]+|\Z)',
    flags=re.DOTALL

)

# 2) Adegan: "S1 — Titik Kumpul di L01", "S2 — Perjalanan ke L02", dst.
#    Mendukung variasi dash: -, –, —
RE_SCENE = re.compile(
    r'(?m)^\s*(S\d{1,2}\s*[—\-–]\s*[^\n]+)\s*\n(.*?)(?=^\s*S\d{1,2}\s*[—\-–]\s*[^\n]+|\Z)',
    flags=re.DOTALL
)

# 3) Q–A: baris yang diawali "Q:" lalu diikuti "A:" (bisa berulang)
RE_QA_PAIR = re.compile(
    r'(?m)^\s*Q\s*:\s*(.+?)\s*\n\s*A\s*:\s*(.+?)(?=\n\s*Q\s*:|\Z)',
    flags=re.DOTALL
)

for path in PDFS:
    print(f"\n=== Proses: {os.path.basename(path)} ===")
    full_text, page_texts = read_pdf_text(path)  # fungsi dari Cell 4
    source_name = os.path.basename(path)
    before_count = len(all_docs)

    # 1) Ekstrak per heading bernomor
    for m in RE_HEADING_NUM.finditer(full_text):
        title = m.group(1).strip()
        body  = (m.group(2) or "").strip()

        # hint halaman: cari kemunculan judul
        page_hint = None
        pat = re.escape(title)
        for (pg, txt) in page_texts:
            if re.search(pat, txt, flags=re.IGNORECASE):
                page_hint = pg
                break

        all_docs.append({
            "text": f"{title}\n{body}",
            "section": title,
            "source": source_name,
            "page_hint": page_hint
        })

    # 2) Ekstrak Adegan Sx — ... (potongan lebih granular)
    for m in RE_SCENE.finditer(full_text):
        title = m.group(1).strip()      # contoh: "S3 — Museum L02 (09:00–10:00)"
        body  = (m.group(2) or "").strip()

        # hint halaman: cukup cari "S\d+"-nya
        page_hint = None
        s_code = re.match(r'^\s*(S\d{1,2})', title)
        if s_code:
            s_pat = re.escape(s_code.group(1))
            for (pg, txt) in page_texts:
                if re.search(rf'\b{s_pat}\b', txt):
                    page_hint = pg
                    break

        all_docs.append({
            "text": f"{title}\n{body}",
            "section": title,            # simpan judul adegan sebagai section
            "source": source_name,
            "page_hint": page_hint
        })

    # 3) Ekstrak Q–A (berguna untuk retrieval pertanyaan langsung)
    for m in RE_QA_PAIR.finditer(full_text):
        q = m.group(1).strip()
        a = m.group(2).strip()
        qa_text = f"Q: {q}\nA: {a}"
        all_docs.append({
            "text": qa_text,
            "section": "Q-A",
            "source": source_name,
            "page_hint": None
        })

    # 4) Fallback: jika tidak ada potongan yang berhasil ditambahkan dari file ini,
    #    gunakan chunking ukuran tetap
    after_count = len(all_docs)
    if after_count == before_count:
        print("Pola heading/adegan/Q-A tidak ditemukan. Menggunakan fallback chunking...")
        step = 1200
        for i in range(0, len(full_text), step):
            chunk = full_text[i:i+step].strip()
            if not chunk:
                continue
            all_docs.append({
                "text": chunk,
                "section": "CHUNK",
                "source": source_name,
                "page_hint": None
            })

print("chunks:", len(all_docs))
```

```python
# Cell 7 - Setup embedding & FAISS index (PDF-only)
from sentence_transformers import SentenceTransformer
import numpy as np
import faiss

if not all_docs:
    raise ValueError("Tidak ada dokumen untuk di-embed. Pastikan Cell 6 menghasilkan 'all_docs' > 0.")

EMB_MODEL = "BAAI/bge-m3"
print(f"Memuat embedding model: {EMB_MODEL}")
embedder = SentenceTransformer(EMB_MODEL)

corpus = [d["text"] for d in all_docs]
print(f"Jumlah chunk untuk di-embed: {len(corpus)}")

# Encode ke vektor (cosine via normalize_embeddings=True → pakai IndexFlatIP)
X = embedder.encode(
    corpus,
    batch_size=64,
    normalize_embeddings=True,
    show_progress_bar=True
)
X = np.array(X, dtype="float32")

print("Membangun FAISS index (Inner Product / Cosine-normalized)...")
index = faiss.IndexFlatIP(X.shape[1])
index.add(X)

print(f"Index siap. Dimensi: {X.shape[1]}, Jumlah vektor: {index.ntotal}")
```

```python
# Cell 8 - Setup reranker (opsional tapi disarankan)
from sentence_transformers import CrossEncoder
import numpy as np

RERANKER = "BAAI/bge-reranker-v2-m3"
print(f"Memuat reranker: {RERANKER}")
reranker = CrossEncoder(RERANKER)  # otomatis pakai GPU jika tersedia (via torch)

def rerank_pairs(query: str, docs: List[Dict], topk: int = 5) -> List[Dict]:
    """
    Re-ranking doc hasil retrieval awal (mis. dari FAISS) dengan CrossEncoder.
    Mengembalikan hingga topk dokumen terbaik, menambahkan 'rerank_score'.
    """
    if not docs:
        return []
    k = min(topk, len(docs))
    pairs = [[query, d["text"]] for d in docs]
    scores = np.asarray(reranker.predict(pairs), dtype="float32")
    order = np.argsort(-scores)[:k]
    return [docs[i] | {"rerank_score": float(scores[i])} for i in order]
```

```python
# Cell 9 - Setup LLM
LLM = "Qwen/Qwen2.5-1.5B-Instruct"

device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained(LLM, use_fast=True)
model = AutoModelForCausalLM.from_pretrained(
    LLM,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    low_cpu_mem_usage=True,
)
model.to(device)
model.eval()

def make_messages(query: str, ctxs: List[Dict]) -> List[Dict]:
    bullets = []
    for c in ctxs:
        snippet = re.sub(r"\s{2,}", " ", c["text"]).strip()[:900]
        bullets.append(f"- {snippet}")
    context_block = "\n".join(bullets) if bullets else "(konteks kosong)"
    return [
        {
            "role": "system",
            "content": (
                "Anda asisten akademik yang menjawab berdasarkan konteks yang diberikan. "
                "Jawab SINGKAT, JELAS, dan HANYA dari konteks. "
                "Jika tidak ada informasi relevan, balas: 'Tidak ditemukan dalam dokumen.' "
                "Jangan sertakan rujukan/sumber dalam jawaban."
            ),
        },
        {
            "role": "user",
            "content": (
                f"KONTEKS:\n{context_block}\n\n"
                f"PERTANYAAN:\n{query}\n\n"
                f"Jawab singkat dan jelas tanpa menyertakan rujukan."
            ),
        },
    ]

@torch.inference_mode()
def generate_answer(messages, max_new_tokens=256):
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    input_len = inputs["input_ids"].shape[1]
    out = model.generate(
        **inputs,
        do_sample=False,
        max_new_tokens=max_new_tokens,
        pad_token_id=tokenizer.eos_token_id,
        eos_token_id=tokenizer.eos_token_id,
    )[0]
    gen = out[input_len:]
    txt = tokenizer.decode(gen, skip_special_tokens=True).strip()
    return re.sub(r"^assistant\s*:\s*", "", txt, flags=re.I) or "Tidak ditemukan dalam dokumen."
```

```python
# Cell 10 - Retrieval & Chat khusus Peraturan Kemahasiswaan ITB 2022

from dataclasses import dataclass
from typing import List, Dict, Optional
import numpy as np
import re, os

@dataclass
class Retrieved:
    text: str
    score: float
    section: str
    source: str
    page_hint: Optional[int]

# --- Booster domain "Peraturan Kemahasiswaan" ---
PAT_PASAL = re.compile(r"\b(Pasal\s?\d{1,2}|Bab\s?[IVXLC]+)\b", re.IGNORECASE)
REG_TERMS = r"(Beasiswa|Asrama|Kesehatan|Konseling|Karakter|Kewirausahaan|Organisasi|Himpunan|UKM|Ganesha|Mapres)"
IMPORTANT_TERMS = rf"({REG_TERMS}|{PAT_PASAL.pattern})"

def _boost_ids_from_query(query: str) -> List[int]:
    """Ambil doc yang mengandung nomor Pasal/Bab atau keyword penting dari query."""
    boost_ids: List[int] = []
    # SAFETY CHECK: Pastikan all_docs sudah ada
    if 'all_docs' not in globals() or not query:
        return boost_ids

    pasal_targets = set(m.group(1).upper() for m in PAT_PASAL.finditer(query))
    keywords = ["BEASISWA", "ASRAMA", "ORGANISASI", "GANESHA", "MAPRES", "IZIN"]

    for i, d in enumerate(all_docs):
        txt = (d.get("text") or "").upper()
        hit = False
        for p in pasal_targets:
            if re.search(rf"\b{re.escape(p)}\b", txt):
                hit = True; break
        if not hit:
            for kw in keywords:
                if kw in query.upper() and kw in txt:
                    hit = True; break
        if hit:
            boost_ids.append(i)
    return boost_ids

def make_messages_reg(query: str, ctxs: List[Dict]) -> List[Dict]:
    """Template pesan untuk LLM yang menjawab Peraturan Kemahasiswaan ITB 2022."""
    bullets = []
    for c in ctxs:
        snippet = re.sub(r"\s{2,}", " ", c["text"]).strip()[:1000]
        bullets.append(f"- {snippet}")
    context_block = "\n".join(bullets) if bullets else "(konteks kosong)"

    return [
        {
            "role": "system",
            "content": (
                "Anda adalah asisten AI resmi untuk Peraturan Kemahasiswaan ITB 2022. "
                "Jawab secara FORMAL dan HANYA berdasarkan konteks. Sebutkan nomor Pasal jika tersedia."
            ),
        },
        {
            "role": "user",
            "content": f"KONTEKS:\n{context_block}\n\nPERTANYAAN:\n{query}",
        },
    ]

def retrieve(query: str, k: int = 15, topk_final: int = 6) -> List[Retrieved]:
    # SAFETY CHECK: Pastikan embedder dan index sudah ada
    if 'embedder' not in globals() or 'index' not in globals():
        raise NameError("Model 'embedder' atau 'index' FAISS belum dimuat. Jalankan Cell sebelumnya!")

    bonus_ids = _boost_ids_from_query(query)
    qv = np.array(embedder.encode([query], normalize_embeddings=True), dtype="float32")
    D, I = index.search(qv, min(k, index.ntotal))
    I = I[0].tolist(); D = D[0].tolist()

    cand, seen = [], set()
    for bid in bonus_ids + I:
        if bid in seen: continue
        seen.add(bid)
        d = all_docs[bid]
        base = D[I.index(bid)] if bid in I else 0.0
        sc = base + (0.15 if bid in bonus_ids else 0.0)
        cand.append(d | {"score": float(sc)})

    # Urutkan berdasarkan skor
    cand.sort(key=lambda x: x["score"], reverse=True)
    out: List[Retrieved] = []
    for c in cand[:topk_final]:
        out.append(Retrieved(
            text=c["text"],
            score=float(c.get("score", 0.0)),
            section=c.get("section", "Peraturan"),
            source=c.get("source", "PDF"),
            page_hint=c.get("page_hint")
        ))
    return out

def chat(query: str) -> Dict:
    try:
        ctxs = retrieve(query)
        msgs = make_messages_reg(query, [c.__dict__ for c in ctxs])
        # Pastikan fungsi generate_answer sudah ada di memori Colab
        ans = generate_answer(msgs)
        cites = [{"section": c.section, "page_hint": c.page_hint, "source": c.source} for c in ctxs]
        return {"answer": ans, "contexts": cites}
    except Exception as e:
        return {"answer": f"Error: {str(e)}", "contexts": []}

print("="*60)
print("RAG Peraturan Kemahasiswaan ITB 2022 — Sistem Siap!")
print("="*60)
```
